```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from tensorflow import random

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
X = pd.read_csv('datasets/HCV_feats.csv')
y = pd.read_csv('datasets/HCV_target.csv')
```

# Modelo 1

```python
seed = 1
np.random.seed(seed)
scaler = StandardScaler()
#X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=seed)
```

```python
np.random.seed(seed)
random.set_seed(seed)

model = Sequential()

model.add(Dense(3, activation='tanh', input_dim=X_train.shape[1]))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='sgd', metrics=['accuracy'], loss='binary_crossentropy')
model.summary()
```

```python
history = model.fit(X_train, y_train, batch_size=20, epochs=100, validation_split=0.2, shuffle=False)
```

```python
plt.figure(figsize=[8, 5])

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('Épocas')
plt.ylabel('Loss')
plt.title('Loss')
plt.legend(['Treino', 'Validação'])

plt.show()
```

```python
print('Melhor acurácia no treino: {} %'.format(round(max(history.history['accuracy']) * 100), 3))
print('Melhor acurácia na validação: {} %'.format(round(max(history.history['val_accuracy']) * 100), 3))

test_loss, test_acc = model.evaluate(X_test, y_test)
print('Melhor loss no teste', round(test_loss, 3))
print('Melhor acurácia no teste', round(test_acc, 3))
```
