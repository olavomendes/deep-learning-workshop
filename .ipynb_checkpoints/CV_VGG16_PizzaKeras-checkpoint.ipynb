```python
import numpy as np

from keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from keras.preprocessing import image
```

```python
classifier = VGG16()
```

```python
new_image = image.load_img('datasets/pizza.jpg', target_size=(224, 224))
new_image
```

```python
new_image = image.img_to_array(new_image)
new_image.shape
```

```python
new_image = np.expand_dims(new_image, axis=0)
new_image.shape
```

```python
new_image = preprocess_input(new_image)
new_image
```

```python
preds = classifier.predict(new_image)
preds.shape
```

```python
decode_predictions(preds, top=5)
```

```python
label = decode_predictions(preds)
decode_label = label[0][0]
print('%s - %.2f%%' % (decode_label[1], decode_label[2] * 100))
```

```python
import numpy as np

from keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from keras.preprocessing import image
```

```python
classifier = VGG16()
```

```python
new_image = image.load_img('datasets/motorcycle.jpg', target_size=(224, 224))
new_image
```

```python
new_image = image.img_to_array(new_image)
new_image = np.expand_dims(new_image, axis=0)
new_image.shape
```

```python
new_image = preprocess_input(new_image)
new_image
```

```python
preds = classifier.predict(new_image)
decode_predictions(preds, top=5)
```

```python
label = decode_predictions(preds)
decode_label = label[0][0]
print('%s - %.2f%%' % (decode_label[1], decode_label[2] * 100))
```
