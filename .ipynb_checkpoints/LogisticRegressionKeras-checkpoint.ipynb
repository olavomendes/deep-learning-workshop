```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from keras.layers import Dense, Activation
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from tensorflow import random
```

```python
feats = pd.read_csv('datasets/OSI_feats.csv')
target = pd.read_csv('datasets/OSI_target.csv')
```

```python
feats.head()
```

```python
feats.drop(['Month', 'VisitorType'], axis=1, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(feats, target, test_size = 0.2, random_state=42)
```

```python
np.random.seed(42)
random.set_seed(42)
```

```python
model = Sequential()
```

```python
model.add(Dense(1, input_dim=X_train.shape[1]))
```

```python
model.add(Activation('sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
print(model.summary())
```

```python
history = model.fit(X_train, y_train['Revenue'], epochs=100, validation_split=0.2, shuffle=False)
```

```python
plt.figure(figsize=[10, 6])

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Acurácia do modelo')
plt.ylabel('Acurácia')
plt.xlabel('Épocas')
plt.legend(['Treino', 'Validação'], loc='lower right')
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Loss')
plt.ylabel('Loss')
plt.xlabel('Épocas')
plt.legend(['Treino', 'Validação'], loc='upper right')

plt.show()
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test['Revenue'])
print(f'The loss on the test set is {test_loss:.4f} and the accuracy is {test_acc*100:.3f}%')
```
